```python
#Which candidate exhibited the most positive dominant emotion (e.g., happy) throughout their video?**
 #  - This can indicate a generally positive attitude, which might be desirable for many roles.

import pandas as pd

# Assuming your DataFrame is named 'df'
df = pd.read_csv('EMOTION_DATA.csv')
# Replace 'df' with the actual variable name of your DataFrame if different

# Group the DataFrame by 'movie_id' to find the candidate with the highest total 'happy' emotion score
happy_scores = df.groupby('movie_id')['happy'].sum()

# Find the movie_id with the highest total 'happy' emotion score
most_positive_movie_id = happy_scores.idxmax()

# Find the candidate (unique ID) associated with the most positive movie_id
most_positive_candidate = df[df['movie_id'] == most_positive_movie_id]['movie_id'].values[0]

print(f"The candidate with the most positive dominant emotion is: {most_positive_candidate}")
```

```python
#candidates who consistently display nervousness or discomfort?
#Excessive nervousness may be a concern, particularly for client-facing roles.

import pandas as pd

# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual variable name of your DataFrame if different

# Calculate the standard deviation of 'fear' emotion score for each candidate
fear_std_dev = df.groupby('movie_id')['fear'].std().reset_index()

# Define a threshold for standard deviation to identify candidates who consistently display fear
threshold_std_dev = 0.5  # You can adjust this threshold as needed

# Filter candidates with low standard deviation (indicating consistency in displaying fear)
consistently_fearful_candidates = fear_std_dev[fear_std_dev['fear'] < threshold_std_dev]

# Print the list of candidates who consistently display fear
if not consistently_fearful_candidates.empty:
    print("Candidates who consistently display nervousness or discomfort:")
    print(consistently_fearful_candidates)
else:
    print("No candidates consistently display nervousness or discomfort.")
```

```python
#Did candidates who blinked more frequently perform differently from those who didn't blink as much?


import pandas as pd
import numpy as np

# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual variable name of your DataFrame if different

# Convert emotion score columns to numeric, handling non-numeric values by replacing with NaN
emotion_columns = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']
df[emotion_columns] = df[emotion_columns].apply(pd.to_numeric, errors='coerce')

# Calculate the average emotion scores for candidates who blinked frequently (blink=1)
blinked_frequently_group = df[df['blink'] == 1].groupby('movie_id')[emotion_columns].mean()

# Calculate the average emotion scores for candidates who didn't blink as much (blink=0)
blinked_less_group = df[df['blink'] == 0].groupby('movie_id')[emotion_columns].mean()

# Print the average emotion scores for both groups
print("Average emotion scores for candidates who blinked frequently (blink=1):")
print(blinked_frequently_group)

print("\nAverage emotion scores for candidates who blinked less (blink=0):")
print(blinked_less_group)



```

```python
 #correlation between the length of the video (elapsed_time) and the emotions expressed?**
  # - Do candidates express more emotions as their video progresses? Does the length of the video impact their performance?

import pandas as pd

# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual variable name of your DataFrame if different

# Calculate the correlation between 'elapsed_time' and each emotion score
emotion_columns = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']
correlation_results = df[['elapsed_time'] + emotion_columns].corr(method='pearson')

# Print the correlation matrix
print("Correlation between 'elapsed_time' and emotion scores:")
print(correlation_results)
```

```python
# Which candidate had the most consistent emotional expression throughout the video?**
 #  - Consistency in emotional expression can be important for roles requiring emotional stability 

import pandas as pd

# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual variable name of your DataFrame if different

# List of emotion columns for which you want to calculate consistency
emotion_columns = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

# Calculate the standard deviation of each emotion score for each candidate
std_deviation_by_candidate = df.groupby('movie_id')[emotion_columns].std()

# Calculate the mean standard deviation for each candidate
mean_std_deviation_by_candidate = std_deviation_by_candidate.mean(axis=1)

# Find the candidate with the lowest mean standard deviation (most consistent)
most_consistent_candidate = mean_std_deviation_by_candidate.idxmin()

print(f"The candidate with the most consistent emotional expression is: {most_consistent_candidate}")
```

```python
#distribution of emotion score
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('EMOTION_DATA.csv')
# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual variable name of your DataFrame if different

# List of emotional score columns
emotion_columns = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

# Plot histograms for each emotional score
plt.figure(figsize=(15, 10))
for emotion in emotion_columns:
    plt.subplot(3, 3, emotion_columns.index(emotion) + 1)
    plt.hist(df[emotion], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {emotion.capitalize()} Emotion')
    plt.xlabel('Emotion Score')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.savefig("distribution of different emotions .jpg")
plt.show()
```

```python
 #What is the distribution of emotional scores (angry, disgust, fear, happy, sad, surprise, neutral) across all candidates? Create histograms or bar plots to visualize these distributions

import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual variable name of your DataFrame if different

# Group the data by 'movie_id'
grouped = df.groupby('movie_id')

# List of emotional score columns
emotion_columns = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

# Create subplots for each emotional score
plt.figure(figsize=(15, 10))

for i, emotion in enumerate(emotion_columns, 1):
    plt.subplot(3, 3, i)
    plt.title(emotion)
    
    # Create histograms for each movie ID
    for movie_id, group_data in grouped:
        plt.hist(group_data[emotion], bins=20, alpha=0.5, label=f'Movie {movie_id}')

    plt.legend()

plt.tight_layout()
plt.show()
```

```python
#distribution of different emotions on different movie id
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('EMOTION_DATA.csv')
# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual variable name of your DataFrame if different

# List of emotional score columns
emotion_columns = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

# Create subplots for each emotion score with adjusted spacing
fig, axes = plt.subplots(nrows=len(emotion_columns), ncols=1, figsize=(10, 20), sharex=True)
plt.subplots_adjust(hspace=0.5)  # Adjust the vertical spacing between subplots

# Iterate through each emotion and create a box plot for each movie_id
for i, emotion in enumerate(emotion_columns):
    sns.boxplot(data=df, x='movie_id', y=emotion, ax=axes[i])
    axes[i].set_title(f'Box Plot of {emotion} by Movie ID')
    axes[i].set_xlabel('Movie ID')
    axes[i].set_ylabel(emotion)
    
    # Decrease font size and rotate x-axis labels
    axes[i].tick_params(axis='x', labelrotation=45, labelsize=8)  # Adjust label rotation and font size

# Adjust the layout and display the plots
plt.tight_layout()
plt.savefig("analysis of different emotions on fifferent movie id.jpg")
plt.show()

```

```python
 #Emotion Trends Over Movies: Do certain movies exhibit specific emotional trends? Create line plots for emotional scores over elapsed_time, grouped by movie_id.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual variable name of your DataFrame if different

# List of emotional score columns
emotion_columns = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

# Create subplots for each emotion score
fig, axes = plt.subplots(nrows=len(emotion_columns), ncols=1, figsize=(10, 20))

# Iterate through each emotion and create line plots for each movie_id
for i, emotion in enumerate(emotion_columns):
    sns.lineplot(data=df, x='elapsed_time', y=emotion, hue='movie_id', ax=axes[i])
    axes[i].set_title(f'Emotion Trend for {emotion}')
    axes[i].set_xlabel('Elapsed Time (seconds)')
    axes[i].set_ylabel(emotion)
    axes[i].legend(title='Movie ID', loc='upper left', bbox_to_anchor=(1, 1))  # Adjust legend position

# Adjust the layout and display the plots
plt.tight_layout()
plt.savefig(" Emotion Trends Over Movies.jpg")
plt.show()
```

```python
# Correlation Matrix: Create a heatmap of the correlation matrix between emotional scores. This can help identify which emotions are strongly correlated or inversely correlated
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('EMOTION_DATA.csv')
# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual variable name of your DataFrame if different

# List of emotional score columns
emotion_columns = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

# Calculate the correlation matrix
correlation_matrix = df[emotion_columns].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap of Emotional Scores')
plt.savefig(" correlation heatmap of emotion scores.jpg")
plt.show()
```

```python
#variation on different emotions on candidates when gaze is 1 or 0
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual variable name of your DataFrame if different

# List of emotional score columns
emotion_columns = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

# Get unique movie IDs from the DataFrame
unique_movie_ids = df['movie_id'].unique()

# Create subplots for each emotion score
fig, axes = plt.subplots(nrows=len(emotion_columns), ncols=len(unique_movie_ids), figsize=(20, 20))

# Iterate through each emotion and each movie ID, and create a box plot for each
for i, emotion in enumerate(emotion_columns):
    for j, movie_id in enumerate(unique_movie_ids):
        data_subset = df[df['movie_id'] == movie_id]
        ax = sns.boxplot(data=data_subset, x='gaze', y=emotion, ax=axes[i, j])
        ax.set_title(f'{emotion}', fontsize=10)  # Display only the emotion name as the title
        ax.set_xlabel('Gaze')  # Add x-axis label 'Gaze'
        ax.set_ylabel(emotion, fontsize=8)  # Decrease y-axis label font size

        # Hide x-axis labels
        ax.set_xticklabels([])

        # Hide y-axis labels except for the leftmost column
        if j > 0:
            ax.set_yticklabels([])

# Adjust the layout with more spacing between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.5)

# Display the plots
plt.savefig("plot the emotion score when gaze is 1 or 0 .jpg")
plt.show()






```
